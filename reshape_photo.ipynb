```python
from PIL import Image
import os
```

```python
def pad_image_to_1000x1000(image, gray):
    width, height = image.size
    if width == 1000 and height == 1000:
        return image
    print('Nie wymiar 1000x1000')
    print(image)
    if gray:
        image = image.convert("L")
        new_image = Image.new("L", (1000, 1000), 0)
        print(new_image.mode)
    else:
        new_image = Image.new("RGB", (1000, 1000), (255, 255, 255))
    left = (1000 - width) // 2
    top = (1000 - height) // 2
    new_image.paste(image, (left, top))
    return new_image

def process_images_in_folder(folder_path, gray=False):
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(folder_path, filename)
            img = Image.open(img_path)
            if img.size != (1000, 1000):
                padded_img = pad_image_to_1000x1000(img, gray)
                padded_img.save(img_path)
                print(f"Zmieniono rozmiar {filename}")
            else:
                print(f"{filename} ma już rozmiar 1000x1000")
```
